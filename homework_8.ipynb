```python
import pandas as pd

district_to_city = pd.read_csv('./materials/hw_8/district_to_city.csv')
city_weather = pd.read_csv('./materials/hw_8/city_weather.csv')
city_weather['average_temperature'] = (city_weather['day_temperatre'] + city_weather['night_temperature']) / 2
average_city_weather = city_weather.groupby('city')['average_temperature'].mean().reset_index()
city_and_district = pd.merge(average_city_weather, district_to_city, on='city')

average_district_weather = city_and_district.groupby('federal_district')['average_temperature'].mean().reset_index()
average_district_weather.rename(columns={'average_temperature': 'average_district_temperature'}, inplace=True)

temperature_data = pd.merge(city_and_district, average_district_weather, on='federal_district')
temperature_data['temperature_difference'] = temperature_data['average_temperature'] - temperature_data['average_district_temperature']
temperature_data.head(10)
```
